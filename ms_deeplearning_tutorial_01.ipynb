```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)

df = shuffle(df)

df.head()
```

```python
X = df.iloc[:, 0:4].values
y = df.iloc[:, 4].values
print(X[0:5])
print(y[0:5])
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data, train_labels, test_labels = train_test_split(X, y, test_size=0.25)

train_labels = np.where(train_labels == 'Iris-setosa', 1, -1)
test_labels = np.where(test_labels == 'Iris-setosa', 1, -1)

print('Train Data\n', train_data[0:2])
print('Trains Labels', np.unique(train_labels))

print('Test Data\n', test_data[0:2])
print('Test Labels', np.unique(test_labels))
```

```python
class Perceptron(object):
  def __init__(self, learning_rate = 0.2, iterations = 10):
    self.learning_rate = learning_rate
    self.n_iterations = iterations

  def weighted_sum(self, X):
    return np.dot(X, self.weights[1:]) + self.weights[0] # a = wx + b

  def predict(self, X):
    return np.where(self.weighted_sum(X) >= 0.0, 1, -1)

  def fit(self, X, y):
    #print(f"DEBUG IN FIT: Type of X: {type(X)}, Shape: {X.shape}")
    #print(f"DEBUG IN FIT: Type of y: {type(y)}, Shape: {y.shape}")

    self.weights = np.zeros(1 + X.shape[1])
    self.errors = []

    #print("Weights", self.weights)

    for _ in range(self.n_iterations):
        error = 0

        for xi, label in zip(X, y): # Renamed inner loop 'y' to 'label'
          y_pred = self.predict(xi)

          update = self.learning_rate * (label - y_pred) # Use 'label' here


          self.weights[1:] = self.weights[1:] + update * xi
          # print("updated weights: ", self.weights[1:]) # Commented out to reduce verbose output during debugging


          self.weights[0] = self.weights[0] + update


          error += int(update != 0.0)

        self.errors.append(error)
```

```python
#from sklearn.linear_model import Perceptron
"""NOTE - this line was causing further code to not use our custom Perceptron rather sklearn library's
pre-built perceptron. See ADDITIONAL TASK at the end for perceptron implementation with sklearn"""
perceptron = Perceptron(learning_rate = 0.1, iterations = 10) # Changed arguments to match custom Perceptron
perceptron.fit(train_data, train_labels)

test_preds = perceptron.predict(test_data)
print("Test Predictions: ", test_preds)
```

```python
from sklearn.metrics import accuracy_score

#y_preds = perceptron.predict(test_data) round(accuracy, 2)
accuracy = accuracy_score(test_preds, test_labels)
print('Accuracy:', accuracy * 100, "%")
```

Task 02<br>Replace the Step Function with Sigmoid.


```python
def predict_with_sigmoid(self, X):

  # First, calculate the weighted sum
  z = self.weighted_sum(X)
  # Apply the sigmoid function to get probabilities between 0 and 1
  sigmoid_output = 1 / (1 + np.exp(-z))
  # Classify based on a threshold (e.g., 0.5), returning 1 or -1
  return np.where(sigmoid_output >= 0.5, 1, -1)


perceptron.predict = predict_with_sigmoid
single_prediction = perceptron.predict(perceptron, np.array(user_input).reshape(1, -1))
print(f"Model's Prediction: {single_prediction} - { 'Iris setosa' if single_prediction == 1 else 'not Iris Setosa' }")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

print("--- scikit-learn Logistic Regression (Perceptron with Sigmoid Activation) Pipeline ---")

# Adjust labels to 1 (for Iris-setosa) and 0 (for not Iris-setosa) for Logistic Regression
train_labels = np.where(train_labels == 1 , 1, 0)
test_labels = np.where(test_labels == 1, 1, 0)

print('Trains Labels', np.unique(train_labels))
print('Test Labels', np.unique(test_labels))

# 1. Feature Scaling (important for Logistic Regression)
scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(train_data)
test_data_scaled = scaler.transform(test_data)

print("\nScaled Training Data (first 2 samples):\n", train_data_scaled[0:2])
print("Scaled Test Data (first 2 samples):\n", test_data_scaled[0:2])

# 2. Initialize and Train Logistic Regression Model
# LogisticRegression uses sigmoid activation by default for binary classification
log_reg_model = LogisticRegression(random_state=42, max_iter=200)
log_reg_model.fit(train_data_scaled, train_labels)

print("\nLogistic Regression Model Trained.")

# 3. Make Predictions on the Test Set
sklearn_test_preds = log_reg_model.predict(test_data_scaled)
sklearn_test_proba = log_reg_model.predict_proba(test_data_scaled) # Probabilities

print("\nTest Predictions (first 5):", sklearn_test_preds[0:5])
print("Test Probabilities (first 5):\n", sklearn_test_proba[0:5])

# 4. Evaluate the Model
sklearn_accuracy = accuracy_score(test_labels, sklearn_test_preds)
print(f"\nscikit-learn Logistic Regression Accuracy: {sklearn_accuracy * 100:.2f}%")

# 5. Predict on Manual User Input
# Ensure user_input is scaled before prediction
user_input_scaled = scaler.transform(np.array(user_input).reshape(1, -1))
single_prediction_sklearn = log_reg_model.predict(user_input_scaled)
single_proba_sklearn = log_reg_model.predict_proba(user_input_scaled)

prediction_label = 'Iris setosa' if single_prediction_sklearn[0] == 1 else 'not Iris Setosa'
print(f"\nModel's Prediction for user input: {single_prediction_sklearn[0]} - {prediction_label}")
print(f"Probability of being Iris Setosa: {single_proba_sklearn[0][1]:.4f}")
```
