```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)

df = shuffle(df)

df.head()
```

```python
X = df.iloc[:, 0:4].values
y = df.iloc[:, 4].values
print(X[0:5])
print(y[0:5])
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data, train_labels, test_labels = train_test_split(X,y, test_size=0.25)

train_labels = np.where(train_labels == 'Iris-setosa', 1, -1)
test_labels = np.where(test_labels == 'Iris-setosa', 1, -1)

print('Train Data', train_data[0:2])
print('Trains Labels', train_labels[0:2])

print('Test Data', test_data[0:2])
print('Test Labels', test_labels[0:2])
```

```python
class Perceptron(object):
  def __init__(self, learning_rate = 0.2, iterations = 10):
    self.learning_rate = learning_rate
    self.n_iterations = iterations

  def weighted_sum(X, self):
    return np.dot(X, self.weights[1:]) + self.weights[0] # a = wx + b

  def predict(self, X):
    return np.where(self.weighted_sum(X) >= 0.0, 1, -1)

  def fit(self, X, y):
    self.weights = np.zeros(1 + X.shape[1])
    self.errors = []

    print("Weights", self.weights)

    for _ in range(self.n_iterations):
        error = 0

        for xi,y in zip(X,y):
          y_pred = self.predict(xi)

          update = self.learning_rate * (y - y_pred)


          self.weights[1:] = self.weights[1:] + update * xi
          print("updated weights: ", self.weights[1:])


          self.weights[0] - self.weights[0] + update


          error += int(update != 0.0)

        self.errors.append(error)
```

```python
from sklearn.linear_model import Perceptron

perceptron = Perceptron(eta0 = 0.1, max_iter = 10)
perceptron.fit(train_data, train_labels)

test_preds = perceptron.predict(test_data)
print(test_preds)

```

```python
from sklearn.metrics import accuracy_score

y_preds = perceptron.predict(test_data)
accuracy = accuracy_score(y_preds, test_labels)
print('Accuracy:', round(accuracy, 2) * 100, "%")
```

```python
user_input = np.zeros([4])
features = [ 'Sepal Length','Sepal Width','Petal Length','Petal Width']
for i in range(4):
  user_input[i] = input(f"Enter value {features[i]}: ")
```

```python
def predict_with_sigmoid(self, X):
  print('used')
  return 1/(1 + np.exp(-self.weighted_sum(X)) >= 0.0, 1,0)
perceptron.predict = predict_with_sigmoid
```

```python
def sigmoid(z):
  print('used')
  return 1/(1 + np.exp(-z))
perceptron.predict = sigmoid
single_prediction = perceptron.predict(np.array(user_input).reshape(-1, 1))
print(single_prediction)
# print(f"Model's Prediction: {single_prediction} - { 'Iris setosa' if single_prediction == 1 else 'not Iris Setosa' }")
```
